```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline

#Styling Block
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18
FIG_SIZE = (12,9) #change figure size throughout for non-faceted figures
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
sns.set_style('white')
cp = sns.color_palette() #color palette to use throughout 

#optional for choropleth maps
SHAPEFILE_PATH = 'ne_10m_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp'
COUNTRY = 'Mexico'


#************* HELPER FUNCTIONS *****************

# Function to compute weighted mean based on load for groupbys
weighted_mean = lambda x: np.average(x, weights=data.loc[x.index, 'developable_load_in_bin_kwh'])

#remove accents from names
import unicodedata
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return u"".join([c for c in nfkd_form if not unicodedata.combining(c)])

#this nifty function maps a dict of strings (keys), and codes (values) using string matching
import difflib
def match_string_to_code(row, str_in, code_dict, manual_dict = None, verbose=False):
    """Accents, spaces, or other differences in the long states names can throw off mapping.
    This function finds the closest string (slow), and just replaces it with the state_id
    
    to be used in a pandas.apply()
    
    str_in: the column of strings
    Code_Dict: (i.e. dict(zip(data['state'], data['state_id'])))
    Manual_Dict: Force a code for a certain string
    Verbose: Whether to print guesses or not
    """
    str_in = row[str_in]
    
    if manual_dict is None:
        manual_dict = {}
    
    if str_in in manual_dict.keys():
        str_in = manual_dict[str_in]
        
    if str_in in code_dict.keys():
        return code_dict[str_in]
    else:
        str_matches = difflib.get_close_matches(str_in, list(code_dict.keys()), cutoff = 0.3, n = 1)
        
        if len(str_matches) > 0:
            if verbose:
                print('Searching for {}, choosing {}'.format(str_in, str_matches[0]))
            return code_dict[str_matches[0]]
        else:
            if verbose:
                print(str_in, 'no matches!')
            return 'missing!' #this will fail

#values to color scale in new column 'color'
def hexcolormapper(df, column, colorscale):
    bins = np.percentile(df[column], np.linspace(0,100,6))
    binned = np.digitize(df[column], bins)
    norm = matplotlib.colors.Normalize(vmin=min(binned), vmax=max(binned), clip=True)
    mapper = plt.cm.ScalarMappable(norm=norm, cmap=colorscale)
    df['color'] = np.vectorize(lambda x: mcolors.to_hex(mapper.to_rgba(x)))(binned)
    return df
```

```python
#import scenarios from agent_outputs csvs 
#CHANGE TO REPRESENT THE SCENARIOS YOU WANT TO RUN
low = pd.read_csv('agent_outputs_low_costs.csv')
mid = pd.read_csv('agent_outputs_mid_costs.csv')
high = pd.read_csv('agent_outputs_high_costs.csv')

#long df with multiple scenarios
low['scenario'] = 'Low'
mid['scenario'] = 'Mid'
high['scenario'] = 'High'
data = pd.concat([low, mid, high], axis='rows')
data = data.reset_index(drop=True)

#drop big columns (i.e. 8760s based on memory usage)
mem = data.memory_usage(deep=True, index=False).to_dict()
p = np.percentile(list(mem.values()), 99) # return 98th percentile, drop above this
keep_columns = [k for k,v in mem.items() if v < p]
data = data[keep_columns]

data['scenario'] = pd.Categorical(data['scenario'], categories = ['Low','Mid','High'], ordered=True)

data = data.rename({'metric_value':'payback_period'}, axis='columns')

data_decades = data.loc[data['year'].isin([2020,2030,2040,2050])]
```

<span style="font-family:Papyrus; font-size:2em;">Technical Potential</span> 

```python
data['tech_potential'] = data['developable_customers_in_bin'] * data['developable_roof_sqft'] * (data['pv_power_density_w_per_sqft']/1e9)

tech_potential = data.groupby(['scenario','year','sector_abbr'], as_index=False)['tech_potential'].sum()


fig, ax = plt.subplots(figsize=FIG_SIZE)
sns.lineplot('year','tech_potential', style='sector_abbr', data=tech_potential, ax=ax)
ax.set_title('Technical Potential by Year')
ax.set_ylabel('Technical PV Potential (GW)')
ax.set_xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) #put legend outside
sns.despine(fig)
```

<span style="font-family:Papyrus; font-size:2em;">Installed Costs</span> 

```python
installed_cost = data.groupby(['scenario','year','sector_abbr'], as_index=False)['pv_price_per_kw'].mean()

fig, ax = plt.subplots(figsize=FIG_SIZE)
sns.lineplot('year','pv_price_per_kw',hue='scenario', style='sector_abbr', data=installed_cost, ax=ax)
ax.set_title('Installed Cost')
ax.set_ylabel('PV Price per kW')
ax.set_xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine(fig)
```

<span style="font-family:Papyrus; font-size:2em;">Economic Metrics</span> 

```python
# Average Payback Period over Time

payback = data.groupby(['scenario','year','sector_abbr'], as_index=False).agg({'payback_period':weighted_mean}) #assuming metric_value = payback period
fig, ax = plt.subplots(figsize=FIG_SIZE)
sns.lineplot('year','payback_period',hue='scenario', ci=None, data=payback, ax=ax)
ax.set_title('Payback Period')
ax.set_ylabel('Weighted Payback Period (Years)')
ax.set_xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

<span style="font-family:Papyrus; font-size:2em;">Economic Potential</span> 

```python
# Economic potential over time. Economic potential is the amount of capacity that could be deployed exceeding the WACC.
# It includes both existing and hypothetical capacity. It changes with year based on the economic factors in that year
# In the calculation below we exploit the fact that agents select the system size that maximizes NPV. If NPV is positive, by 
# definition that means that the WACC has been exceeded, and if NPV is negative for all system sizes then pv_kw will equal zero.

data['economic_potential_MW'] = data['developable_customers_in_bin'] * data['pv_kw'] / 1000
economic_potential = data.groupby(['scenario','year','sector_abbr'], as_index=False)['economic_potential_MW'].sum()
fig, ax = plt.subplots(figsize=FIG_SIZE)
sns.lineplot('year','economic_potential_MW',hue='scenario', ci=None, data=economic_potential, ax=ax)
ax.set_title('Economic Potential')
ax.set_ylabel('PV MW * Developable Customers')
ax.set_xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine(fig)
```

```python
# Histogram of payback by sector, weighed by developable load

payback = data_decades.groupby(['control_reg','sector_abbr','scenario','year'], as_index=False).agg({'payback_period':weighted_mean})

fg = sns.FacetGrid(payback, col='scenario', row = 'year', hue = 'sector_abbr', margin_titles=True,
                   despine=True, legend_out=True, height=(FIG_SIZE[1]/4), aspect = 1.5)

fg.map(plt.hist, 'payback_period', histtype="step", linewidth=3)
plt.subplots_adjust(top=0.9)
fg.fig.suptitle('Histograms of Control Regions by Payback Period (Weighted by Load)')
```

<span style="font-family:Papyrus; font-size:2em;">Market Share</span> 

```python
market_share = data_decades[['control_reg','sector_abbr','pv_kw_cum','market_share','scenario','developable_load_in_bin_kwh','year']]

market_share = market_share.groupby(['control_reg','sector_abbr','year','scenario'], as_index=False).agg({'market_share':weighted_mean})
market_share = market_share.loc[market_share['market_share'] != np.nan]

sns.catplot(x='sector_abbr',y='market_share',col='scenario',data=market_share, kind='swarm', hue='year',
            height = FIG_SIZE[1], aspect = 0.7)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x) #no scientific notation

deployment = data.groupby(['scenario','year','sector_abbr'], as_index=False)['pv_kw_cum'].sum()
deployment_2050 = deployment.loc[deployment['year']==2050]
deployment_2050
```

```python
import matplotlib.patches as mpatches

sector_abbr_dict = {'res': cp[0],
                    'com': cp[1],
                    'ind': cp[2]}
    
supply_curve_2050 = data.loc[(data['year'] == 2050)]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, figsize=(FIG_SIZE[0]*1,FIG_SIZE[1]), sharex=True, sharey=True)

for s, a in zip(['Low','Mid','High'], [ax1, ax2, ax3]):
    supply_curve = supply_curve_2050.loc[supply_curve_2050['scenario'] == s]
    supply_curve_grouped = supply_curve.groupby(['sector_abbr','control_reg_id'], as_index=False).agg({'payback_period':weighted_mean, 'pv_kw_cum':'sum'})
    supply_curve_grouped = supply_curve_grouped.sort_values('payback_period',ascending=True)

    supply_curve_grouped = supply_curve_grouped.loc[supply_curve_grouped['pv_kw_cum'] > 0]

    width = [float(i / 1000000) for i in supply_curve_grouped['pv_kw_cum']] #to GW
    colors = supply_curve_grouped['sector_abbr'].map(sector_abbr_dict)

    relative_positions = []
    relative_position = 0
    previous_width = 0
    for w in width:
        relative_position = float(relative_position + (previous_width/2) + (w/2))
        previous_width = w
        relative_positions.append(relative_position)

    a.bar(relative_positions, supply_curve_grouped['payback_period'], width=width, linewidth=0.0,
            color=colors)
    a.set_title(s)
 
plt.suptitle('Supply Curves of 2050 Capacity by Pay Back Period and Sector')
plt.subplots_adjust(hspace=0.5)
plt.ylabel('Payback Period')
plt.xlabel('Installed Capacity (GW)')
custom_patches = [mpatches.Patch(color=v, label=k) for k,v in sector_abbr_dict.items()]
plt.legend(handles=custom_patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine(fig)
```

```python
data['pv_MW_cum'] = data['pv_kw_cum'] / 1000
deployment_no_sec = data.groupby(['scenario','year'], as_index=False)['pv_MW_cum'].sum()
fig, ax = plt.subplots(figsize=FIG_SIZE)
sns.lineplot('year','pv_MW_cum',hue='scenario', data=deployment_no_sec, ax=ax)
ax.set_title('Cumulative DPV Capacity')
ax.set_ylabel('MW')
ax.set_xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine(fig)
```

<span style="font-family:Papyrus; font-size:2em;">Choropleth</span> 

```python
#this cell merges a df with a geodataframe, it can be slow
import geopandas as gpd

#shape file specific cleanup
#CHANGE THIS FOR YOUR CASE
MAP_COLUMN = 'pv_kw_cum'

grouped_data = data_decades.groupby(['state_id','year','scenario'], as_index=False)[[MAP_COLUMN]].sum()
# grouped_data = grouped_data.rename({'departamento':'Departamento'}, axis = 'columns')

state_id_map = dict(zip(data['state'], data['state_id']))

shape_in = gpd.read_file(SHAPEFILE_PATH)
shape_in = shape_in.loc[shape_in['admin'] == COUNTRY]
shape_in = shape_in.dropna(subset=['name'])

#search data for matching state names in the shapefile
shape_in['state'] = shape_in['name'].apply(remove_accents)
shape_in['state_id'] = shape_in.apply(match_string_to_code, args=('state', state_id_map), axis=1)
shape_in = shape_in[['state_id','geometry']]

#merge data and shapefile
gdf = shape_in.merge(grouped_data, on = ['state_id'], how = 'inner')
gdf = gdf.dropna(subset=[MAP_COLUMN])

import matplotlib.colors as mcolors
from geopandas.plotting import plot_polygon_collection

gdf = hexcolormapper(gdf, MAP_COLUMN, 'Greens')
gdf = gdf[['geometry','color','pv_kw_cum','scenario','year','state_id']]

#plotting in a loop
for s in ['Low','Mid','High']:
    scale = 8
    fig, (ax20, ax30, ax40, ax50) = plt.subplots(ncols=4,
                                                 sharex=True, sharey=True) #for three scenarios
    df_2020 = gdf.loc[(gdf['year'] == 2020) & (gdf['scenario']==s)]
    df_2030 = gdf.loc[(gdf['year'] == 2030) & (gdf['scenario']==s)]
    df_2040 = gdf.loc[(gdf['year'] == 2040) & (gdf['scenario']==s)]
    df_2050 = gdf.loc[(gdf['year'] == 2050) & (gdf['scenario']==s)]
    
    plot_polygon_collection(ax20, geoms=df_2020['geometry'], facecolor=df_2020['color'], edgecolor ='k', linewidth=0.18, alpha = 0.6)
    plot_polygon_collection(ax30, geoms=df_2030['geometry'], facecolor=df_2030['color'], edgecolor ='k', linewidth=0.18, alpha = 0.6)
    plot_polygon_collection(ax40, geoms=df_2040['geometry'], facecolor=df_2040['color'], edgecolor ='k', linewidth=0.18, alpha = 0.6)
    plot_polygon_collection(ax50, geoms=df_2050['geometry'], facecolor=df_2050['color'], edgecolor ='k', linewidth=0.18, alpha = 0.6)
    
    ax20.set_title('2020')
    ax30.set_title('2030')
    ax40.set_title('2040')
    ax50.set_title('2050')
    
    ax20.axis('off')
    ax30.axis('off')
    ax40.axis('off')
    ax50.axis('off')

    plt.suptitle(s)
    plt.subplots_adjust(wspace=0, hspace=0)
    
    w, h = fig.get_size_inches()
    fig.set_size_inches(w * 4, h)
```

```python
pair_columns = ['scenario','payback_period','market_share','tech_potential','pv_MW_cum','npv']
pair_df = data.loc[data['year'] == 2050]
pair_df = pair_df[pair_columns]

#fillna
pair_df[['payback_period','market_share','tech_potential','pv_MW_cum','npv']] = pair_df[['payback_period','market_share','tech_potential','pv_MW_cum','npv']].fillna(0)

size = np.sqrt(pair_df['pv_MW_cum']) * 5

fg = sns.PairGrid(pair_df, hue='scenario')
fg = fg.map_diag(sns.distplot)
fg = fg.map_upper(plt.scatter, s=size, alpha = 0.2) #upper size is variable based on pv_MW_cum
fg = fg.map_lower(plt.scatter, s=20, alpha = 0.2) #lower size is fixed
fg = fg.add_legend()

plt.subplots_adjust(top=0.9)
fg.fig.suptitle('Pair Plots for EDA (Diaganol is Histogram; Upper Half Size is based on pv_MW_cum)')

fg.axes[0,0].set_ylim(0,22)
fg.axes[0,0].set_xlim(0,22)
fg.axes[1,1].set_ylim(0,1)
fg.axes[1,1].set_xlim(0,1)
fg.axes[2,2].set_ylim(0,5)
fg.axes[2,2].set_xlim(0,5)
fg.axes[3,3].set_ylim(0,1000)
fg.axes[3,3].set_xlim(0,1000)
fg.axes[4,4].set_ylim(0,100000)
fg.axes[4,4].set_xlim(0,100000)
```
